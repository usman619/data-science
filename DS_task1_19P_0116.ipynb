```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

site = "http://www.values.com/inspirational-quotes"
hdr = {'User-Agent': 'Mozilla/5.0'}
request = Request(site, headers = hdr)
page = urlopen(request)
data = BeautifulSoup(page, 'html.parser')
print(data)


```

```python
image = data.find_all('img', attrs={'class', 'margin-10px-bottom shadow'})
for item in image:
  print(item['src'])
```

```python
import pandas as pd

scrapedData = []

image = data.find_all('img', attrs={'class', 'margin-10px-bottom shadow'})

for item in image:
  dicty = {}

  imgLink = item['src']
  quotes,authors= item['alt'].split("#",1)
  dicty ['quotes'] = quotes
  dicty ['authors'] = authors
  dicty ['image link'] = imgLink

  # link = data.find('a')
  # dicty ['quotes link'] = link.get('href')

  scrapedData.append(dicty)

# for link in data.find_all('h2',attrs={'href', 'value-on-red'}):
#  dicty ={}
#  quotesLink = link.get('href')
#  dicty['quotes link'] = quotesLink
#  scrapedData.append(dicty)

print(scrapedData)

dataFrame = pd.DataFrame.from_dict(scrapedData)

```
